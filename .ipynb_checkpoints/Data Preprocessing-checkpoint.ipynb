```python
#standardization: zscore scaling 
#x_stand = (x - mean(x))/standarddeviation
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
data_stand = sc.fit_transform(data)
```

```python
"""
Multi Linear Regression model for Salary prediction
Salary(Y) = b + a1*Gender + a2*Age + a3*WorkingYear + a4*TrainedYear
"""
X = df[['Skill', 'Gender', 'Age', 'WorkingYear', 'TrainedYear']] #indepent variable
Y = df['Salary'] #dependent variable
```

```python
#Encode categorical(dummy) variable Gender (Male:1, Female:0)
#drop_first=True is important to use, as it helps in reducing the extra column created during dummy variable creation. 
#Hence it reduces the correlations created among dummy variables.
X = pd.get_dummies(X, columns = ['Gender'], drop_first=True)# if not have drop_first => 2 columns will be created
X
```

```python
#other ways
#encode dummy variable Gender (Male = 1, Female = 0)
#values for Gender: Male = 1/ Female = 0
df['Gender_new'] = df['Gender'].map({'Male': 1, 'Female': 0})
```

```python
X = df[['Skill', 'Gender_new', 'Age', 'WorkingYear', 'TrainedYear']]
Y = df['Salary']

X = pd.get_dummies(X, columns = ['Gender_new'], drop_first=True)
```

```python
X
```

```python
X['Skill'].unique()
```

```python
X = pd.get_dummies(X, columns = ['Skill'])
```

```python
X
```

```python
#save to excel file
dummy_data = X.join(Y)
dummy_data.to_excel('employee_dummy.xlsx')
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X,Y)
```

```python
print('R square:', lm.score(X, Y))
print('Slope:', lm.coef_)
print('Intercept:', lm.intercept_)
```

```python
pd.DataFrame(lm.coef_, X.columns)
```

```python
#Discretization using cut method of pandas package
df['Working Level'] = pd.cut(df['WorkingYear'], 4, labels = ['Internship', 'Fresher', 'Junior', 'Senior'])
df.head(4)
```

```python
cut_labels = ['Low', 'Medium', 'High']
cut_bins = [0, 800, 1200, 5000]
df['Salary Level'] = pd.cut(df['Salary'], bins = cut_bins, labels = cut_labels)
df.head(10)
```

```python
cut_TrainedYear_labels = ['College', 'University', 'Master_PhD']
cut_TrainedYear_bins = [0, 4, 6, 11]
df['TrainedYear Level'] = pd.cut(df['TrainedYear'], bins = cut_TrainedYear_bins, labels = cut_TrainedYear_labels)
df
```
